```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
from scipy.integrate import solve_ivp

tsim = np.linspace(0,40,200)

def pendulam(t,x,b,c):
    theta,omega = x 
    u = .2*np.sin(2*t)
    return [omega+.1,-b*omega - c*np.sin(theta) + u]

x_init =[0.5*np.pi,0]

sol = solve_ivp(pendulam,[tsim[0], tsim[-1]],x_init,args=(0.25,0.5),t_eval=tsim)

plt.plot(tsim,sol.y[0],label='Theta')
plt.plot(tsim,sol.y[1],label='Omega')
plt.xlabel('Time')
plt.ylabel('States')
plt.title('Simple Pendulam')
plt.legend(loc='best')
plt.show()
```
